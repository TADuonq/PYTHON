### Pylot API vs Object-Oriented(OO) API
* Pylot API -> Quickly
* OO API -> Advanced

#### Pylot API

```python
x = np.linspace(0,10,1000)
plt.plot(x, np.sin(x), color = "blue", linestyle = "dashed", label = "sin(x)");
plt.plot(x, np.cos(x), color = "red", label = "cos(x)");

plt.title("Sin(X) and Cos(X)")
plt.xlabel("Bien X")
plt.ylabel("F(X)")

#plt.xlim([0,4])
#plt.axis([0, 4, -0.7, 0.6])
#plt.axis("tight")
plt.axis("equal")

plt.legend();
```

```python
#1. Prepare data
x = [1,2,3,4]
y = [11,22,33,44]

#2. Setup plot
fig, ax = plt.subplots(figsize =(5,5))

#3. Plot data
ax.plot(x,y);
ax.set(title = "A simple plot",
        xlabel = "x-axis",
        ylabel = "y-axis");
```

## 2. Most common types of Matplotlib plots
* line
* scatter
* bar
* hist
* subplots()

```python
x = np.linspace(0,10,100)
x[:5]
```

```python
fig, ax = plt.subplots()
ax.plot(x, x**3)
```

### Scatter


```python
# Pylot API
plt.scatter(x, np.exp(x));
```

```python
# OO API
rng = np.random.RandomState(0)

x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000*rng.rand(100)

fig, ax = plt.subplots()
img1 = ax.scatter(x, y,s = sizes, c = colors, cmap = 'viridis', alpha = 0.3);

fig.colorbar(img1)
```

### 3.Bar
* Vertical
* Horizontal


```python
soft_drink_prices = {"Coke": 10, "Pepsi": 12, "Sprite" : 8}
fig, ax = plt.subplots()
ax.bar(soft_drink_prices.keys(), soft_drink_prices.values())
ax.set(title = "Bach hoa xanh's Soft Drink Price", ylabel = "Price($)")
```

```python
fig, ax = plt.subplots()
ax.barh(list(soft_drink_prices.keys()), list(soft_drink_prices.values()));
```

### 4.Histogram

```python
# Prepare data 
np.random.seed(42)
student_height = np.random.normal(170, 10, 250)
```

```python
student_height[:10], len(student_height)
```

```python
# pylot API
plt.hist(student_height, bins = 30);
```

### Subplots

```python
# Option l:
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2, ncols = 2, figsize =(10, 5))
# Plot data to each axis
ax1.plot(x, x/2);
ax1.plot(x, x*2);

ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(soft_drink_prices.keys(), soft_drink_prices.values());
ax4.hist(student_height, bins = 30);
```

```python
# Option 2: 
fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize =(10,5))
ax[0,0].plot(x, x/2);
ax[0,0].plot(x, x*2);

ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(soft_drink_prices.keys(), soft_drink_prices.values());
ax[1,1].hist(student_height, bins = 30);
```
